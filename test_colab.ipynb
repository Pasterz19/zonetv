# 🚀 TEST POŁĄCZENIA Z COLAB
Ten notebook sprawdza połączenie z serwerami Google Colab i wyświetla informacje o środowisku.

```python
import sys
import platform
import psutil
import subprocess
from datetime import datetime

print("=" * 60)
print("📡 INFORMACJE O POŁĄCZENIU")
print("=" * 60)
print(f"Data i czas: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Python wersja: {sys.version}")
print(f"Platforma: {platform.system()} {platform.release()}")
print(f"Hostname: {platform.node()}")
```

```python
print("\n" + "=" * 60)
print("💻 SPECYFIKACJA SERWERA")
print("=" * 60)

# CPU
print(f"CPU: {psutil.cpu_count()} rdzeni logicznych")
print(f"CPU freq: {psutil.cpu_freq().current:.0f} MHz" if psutil.cpu_freq() else "CPU freq: brak danych")

# RAM
ram = psutil.virtual_memory()
print(f"RAM: {ram.total / 1024**3:.2f} GB")
print(f"RAM dostępne: {ram.available / 1024**3:.2f} GB")

# Dysk
disk = psutil.disk_usage('/')
print(f"Dysk: {disk.total / 1024**3:.2f} GB")
print(f"Dysk wolne: {disk.free / 1024**3:.2f} GB")
```

```python
print("\n" + "=" * 60)
print("🎮 SPRAWDZANIE GPU")
print("=" * 60)

try:
    gpu = subprocess.check_output(
        ['nvidia-smi', '--query-gpu=name,memory.total', '--format=csv,noheader'],
        text=True
    ).strip()
    if gpu:
        print(f"✅ GPU: {gpu}")
    else:
        print("❌ Brak GPU (tryb CPU)")
except:
    print("❌ Brak GPU (tryb CPU)")
```

```python
print("\n" + "=" * 60)
print("⚡ TEST WYDAJNOŚCI")
print("=" * 60)

import numpy as np
import time

size = 3000
print(f"Test: mnożenie macierzy {size}x{size}...")

start = time.time()
a = np.random.rand(size, size)
b = np.random.rand(size, size)
c = np.dot(a, b)
end = time.time()

print(f"✅ Czas: {end - start:.2f} sekund")
```
